### Metoda `agg` (agregare)
Această metodă are o sintaxă particulară, fiind necesar să transmitem ca argument o structură de tip dicționar.<br>
De exemplu, se cere să se afișeze suma vânzărilor. Vom folosi `sum`, dar îl vom transmite ca valoare a cheii `Sales` a dicționarului. 

```python
df.agg({'Sales':'sum'}).show()
```

Să se determine, în mod similar, valoarea maximă a vânzărilor.

### Combinarea metodelor `groupBy` și `agg` pentru a obține informații despre companie

```python
group_data = df.groupBy('Company')
```

```python
group_data.agg({'Sales':'max'}).show()
```

### Importul și utilizarea altor funcții

```python
from pyspark.sql.functions import stddev,countDistinct,count,avg
```

```python
df.agg({'Sales':'stddev'}).show()
```

#### Rezultatul anterior poate fi obținut și prin utilizarea metodei `select`

```python
df.select(stddev('Sales')).show()
```

#### Să se obțină numărul de vânzări și numărul de vânzări cu valori distincte.

### Formatarea rezultatelor
Multe dintre aceste funcții și operații matematice produc rezultate de tip long number. Putem modifica formatul cu ajutorul funcției `format_number`. De asemenea, putem modifica numele coloanei de tip long cu ajutorul metodei `alias`.

```python
from pyspark.sql.functions import format_number
```

```python
sales_std = df.select(stddev('Sales').alias('std'))
```

```python
sales_std
```

```python
#Să se formateze coloana std ca număr cu 2 zecimale. Rezultatul va avea alias-ul "final".
sales_std=sales_std.select(format_number('std',2).alias('final'))
```

```python
sales_std
```

```python
sales_std.show()
```

### Utilizarea metodei `orderBy` pentru a ordona după valorile unei coloane

```python
df.show()
```

```python
df.orderBy('Sales').show()
```

#### Pentru a ordona descrescător, trebuie să se transmită obiectul coloană (nu doar numele său, ca string) și să se apeleze metoda `desc` asupra lui.

```python
df.orderBy(df['Sales'].desc()).show()
```

Să se afle compania având cea mmai mare valoare totală a vânzărilor.
