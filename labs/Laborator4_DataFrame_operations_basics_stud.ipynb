### Operații de filtrare
Filtrarea se poate realiza cu ajutorul sintaxei SQL-like sau în manieră Python. Vom utiliza ambele moduri.

#### Filtrați liniile a căror valoare de închidere a fost mai mică decât 500 (specificați condiția SQL în metoda `filter()`). Afișati primele 5 linii.

```python
df.filter("Close < 500").show(5)
```

#### Modificați comanda anterioară astfel încât să se obțină doar coloanele `Date`, `Open`, `Close`. Se vor afișa primele 7 linii.

```python
df.filter("Close < 500").select(['Date','Open','Close']).show(7)
```

```python
#Filtrati liniile pentru care deschiderea a fost > 500 iar inchiderea < 500. Afișati primele 5 linii.
df.filter("Close < 500 AND Open > 500").show(5)
```

### Utilizați sintaxa DataFrame pentru cerința inițială: Filtrați datele a căror valoare de închidere a fost mai mică decât 500; afișati primele 5 linii.

```python
df.filter(df['Close']<500).show(5)
```

#### Dacă este necesar să înlănțuim mai multe condiții, folosim `&` pentru AND și `|` pentru OR. Condițiile se separă cu ajutorul parantezelor.

```python
# Utilizând sintaxa DataFrame, filtrați liniile pentru care valoarea de deschidere a fost > 500, iar valoarea de închidere < 500. Afișați primele 5 linii.
df.filter((df['Close']<500) & (df['Open']>500)).show(5)
```

#### Pentru a testa egalitatea, folosim operatorul de comparare `==`. Operatorul NOT este `~`

```python
#Să se afișeze linia pentru care coloana Low are valoarea 197.16 (sintaxa DataFrame)
df.filter(df['Low']==197.16).show()
```
