## Ștergerea datelor lipsă

Putem folosi funcțiile .na pentru datele absente. Comanda drop are următorii parametri:

    df.na.drop(how='any', thresh=None, subset=None)
    
    * param how: 'any' or 'all'.
    
        If 'any', drop a row if it contains any nulls.
        If 'all', drop a row only if all its values are null.
    
    * param thresh: int, default None
    
        If specified, drop rows that have less than `thresh` non-null values.
        This overwrites the `how` parameter.
        
    * param subset: 
        optional list of column names to consider.

```python
# Să se șteargă orice linie de pe care lipsesc date
df.na.drop().show()
```

```python
# Să se șteargă liniile de pe care lipsesc date, păstrându-se liniile pe care există cel puțin 2 valori NOT null.
df.na.drop(thresh=2).show()
```

```python
#Să se șteargă liniile care au valoarea NULL pe coloana Sales.
df.na.drop(subset=["Sales"]).show()
```

```python
#Să se șteargă liniile ce conțin cel puțin o valoare NULL
df.na.drop(how='any').show()
```

```python
#Să se șteargă liniile pe care toate valorile sunt NULL
df.na.drop(how='all').show()
```

## Completarea valorilor NULL

Valorile absente se pot completa cu alte valori. Dacă avem multe valori NULL pentru mai multe tipuri de date, Spark va completa NULL-urile corespunzătoare tipurilor respective.

```python
# Să se completeze valorile NULL corespunzătoare coloanelor de tip String cu valoarea 'NEW VALUE'
df.na.fill('NEW VALUE').show()
```

```python
# Să se completeze valorile NULL corespunzătoare coloanelor de tip numeric cu valoarea 0
df.na.fill(0).show()
```

Se pot specifica doar anumite coloane ce vor fi completate, cu ajutorul parametrului subset.

```python
#Să se completeze valorile NULL ale coloanei Name cu valoarea 'No name'.
df.na.fill('No Name',subset=['Name']).show()
```
