### Crearea de noi coloane (după procesarea coloanelor existente)

#### Următoarea operație din Pandas nu funcționează în Spark, ci va produce o eroare.

```python
df['newage']=2*df['age']
```

#### Se va utiliza metoda `withColumn()`

```python
df.withColumn('double_age',df['age']*2).show()
```

#### Pentru redenumire, se va utiliza metoda `withColumnRenamed()`

```python
df.withColumnRenamed('age','my_new_age').show()
```

#### Se pot face operații cu coloane multiple, precum o sumă de vectori

```python
df2=df.withColumn('half_age',df['age']/2)
df2.show()
```

```python
df2=df2.withColumn('new_age',df2['age']+df2['half_age'])
df2.show()
```

#### La afișarea schemei, vom observa că tipurile de date ale lui `half_age` și `new_age` sunt considerate automat `double` (din cauza operației în virgula mobilă efectuate)
